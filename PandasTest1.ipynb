```python
# import modules
import matplotlib.pyplot as plt
import pandas as pd

plt.close("all")
```

```python
# Load data
fileName = pd.ExcelFile ('Trend Trading Input Output Data.xlsx')
priceData = pd.read_excel(fileName,'Price Data')

print(priceData.head)

# print(priceData.tail)
```

```python
varNames = priceData.columns
print(varNames)
```

```python
timeIndex = priceData.index
print(timeIndex)
```

```python
timeColumn = priceData.Time
print(timeColumn)
```

```python
priceTT = priceData
priceTT.index = timeColumn
priceTT.index

priceTT = priceTT.drop(['Time'], axis=1)
priceTT
```

```python
priceTT
```

```python
priceTT.index
```

```python
varNames =  priceTT.columns
print(varNames)
```

```python
priceTT.dtypes
```

```python
priceTT =  priceTT[varNames].astype(float)
priceTT.dtypes
```

```python
priceTT.head()
```

```python
priceTTclean = priceTT.fillna(method='ffill')
priceTTclean = priceTTclean.fillna(method='bfill')
priceTTclean
```

```python
# plotting
df = priceTTclean
df.values[:,:] = priceTTclean.values/priceTTclean.values[0,:]
df
```

```python
plt.figure()
df.plot()
```

```python
plt.figure()
symbol = 'TKIM'
df[symbol].plot(xlabel = 'Time', ylabel = 'Cumsum' )
```

```python
plt.figure()
symbol = 'ARTO'
df[symbol].plot(xlabel = 'Time', ylabel = 'Cumsum', logy = True )
```

```python
plt.figure()
symbol = 'ARTO'
df[symbol].plot(xlabel = 'Time', ylabel = 'Cumsum', logy = False)
```

```python
# compute daily returns
priceTTclean = priceTT.fillna(method='ffill')
priceTTclean = priceTTclean.fillna(method='bfill')
priceTTclean


dailyReturns = priceTTclean
dailyReturns = priceTTclean.pct_change()
dailyReturns
```

```python
fileName = 'Trend Trading Input Output Data.xlsx'
sheetNames = 'Daily returns'
dailyReturns.to_excel(fileName, sheet_name=sheetNames)
```

```python
plt.figure()
dailyReturns.plot()
```
